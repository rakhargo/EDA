```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_excel("C:\College\Semester 3\Data Mining and Warehouse\provider.xlsx")
data[::]
```

```python
data.loc[data['keberadaan_operator_lain'] == 'TIDAK ADA', 'operator_lainnya'] = 'TIDAK ADA'
# mengecek nilai pada kolom "keberadaan_operator_lainnya". Jika bernilai "TIDAK ADA" maka pada kolom "operator_lainnya" diganti menjadi "TIDAK ADA"
data['operator_lainnya'] = data['operator_lainnya'].replace("0", "TIDAK ADA")
data.head(50)
# Mengecek kolom operator_lainnya. Jika value pada kolom "keberadaaan_operator_lainnya" bernilai "TIDAK ADA" 
# maka value yang semula bernilai 0 diganti dengan "TIDAK ADA"
```

```python
data['jarak_tower_terdekat'] =  data['jarak_tower_terdekat'].apply(lambda x: 1 if x < 1 else x)
# mengubah nilai pada kolom "jarak_tower_terdekat" yang memiliki nilai < 1 untuk diubah semuanya menjadi 1
data.head(50)
```

```python
data = data.astype({'jarak_tower_terdekat': int})
# mengubah tipe data kolom "jarak_tower_terdekat" menjadi integer
print(data.dtypes)
```
